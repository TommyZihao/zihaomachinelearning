### 7.取出关键时间信息
* .dt.month
* .dt.dayofweek
* .dt.dayofyear
* ...

```python
# 取出几月份
car_sales.loc[:,'month'] = car_sales['date'].dt.month
```

```python
car_sales.head()
```

```python
tmp_date = car_sales['date'].dt
```

```python
# 取出来是几号
car_sales.loc[:,'dom'] = car_sales['date'].dt.day
```

```python
# 取出一年当中的第几天
car_sales.loc[:,'doy'] = car_sales['date'].dt.dayofyear
```

```python
# 取出星期几
car_sales.loc[:,'dow'] = car_sales['date'].dt.dayofweek
```

```python
car_sales.head()
```

```python
car_sales.loc[:,'is_weekend'] = car_sales['dow'].apply(lambda x: 1 if (x==0 or x==6) else 0)
```

```python
car_sales.head()
```

## 特征选择

### 过滤式/Filter
* SelectKBest

```python
from sklearn.feature_selection import SelectKBest
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X, y = iris.data, iris.target
```

```python
X.shape
```

```python
X[:5,:]
```

```python
y
```

```python
X_new = SelectKBest(k=2).fit_transform(X,y)
```

```python
X_new.shape
```

```python
X_new[:5,:]
```

### 包裹型/wrapper
* RFE

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rfe = RFE(estimator=rf, n_features_to_select=2)
```

```python
X_rfe = rfe.fit_transform(X,y)
```

```python
X_rfe.shape
```

```python
X_rfe[:5,:]
```

### 嵌入式/Embedded
* SelectFromModel

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.svm import LinearSVC
```

```python
lsvc = LinearSVC(C=0.01, penalty="l1", dual=False).fit(X,y)
```

```python
model = SelectFromModel(lsvc, prefit=True)
```

```python
X_embed = model.transform(X)
```

```python
X_embed.shape
```

```python
X_embed[:5,:]
```
